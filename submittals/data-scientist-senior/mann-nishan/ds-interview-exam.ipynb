### Q 1.1: Fibionacci
![fib image](https://upload.wikimedia.org/wikipedia/commons/thumb/9/93/Fibonacci_spiral_34.svg/200px-Fibonacci_spiral_34.svg.png)

#### Q 1.1.1
Given $n$ where $n \in \mathbb{N}$ (i.e., $n$ is an integer and $n > 0$), write a function `fibonacci(n)` that computes the Fibonacci number $F_n$, where $F_n$ is defined by the recurrence relation:

$$ F_n = F_{n-1} + F_{n-2}$$

with initial conditions of:

$$ F_1 = 1,  F_2 = 1$$

```python
def fibonacci(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
[fibonacci(i) for i in range(1, 11)]
```

#### Q 1.1.2
What's the complexity of your implementation?

$T(n) = \mathcal{O}(2^n)$ where $T$ represents the runtime complexity and $n$ is the nth fibonacci number in the sequence. This was derived by drawing a recursion tree. The time complexity can be written using the recursion relation $$T(n) = T(n) + T(n-1) + \mathcal{O}(1)$$ where $\mathcal{O}(1)$ is the cost of doing of either doing addition or returning some value at each node of the recursion tree. For example, the recursion tree for $T(3)$ looks as follows

        ùí™(1)
        / \
       /   \ 
     ùí™(1)  ùí™(1)

The total cost can be obtained by summing up costs across a level and then summing the level costs over the height of the tree. In this case $T(3) = 3\mathcal{O}(1) = \sum_{i=1}^{2} 2^{i-1}\mathcal{O}(1) $. However drawing the recursion tree for $T(4)$ or $T(8)$, one begins to see the redundant number of computations performed. For the _nth_ Fibonacci number, the tree's depth is $n-1$ and the cost at each depth level $i$, denoted as $\alpha_i$, is less than or equal to by $2^{i-1}\mathcal{O}(1)$. We sum along the height of the tree
$$T(n) = \mathcal{O}\left(\sum_{i=1}^{n-1} \alpha_i\right) < 
   \mathcal{O}\left(\sum_{i=1}^{n-1} 2^{i-1} \right) <
   \mathcal{O}\left(\sum_{i=1}^{n-1} 2^{n-2} \right) = (n-1)2^{n-2} = \mathcal{O}(2^n) $$

#### Q 1.1.3
Consider an alternative implementation to compute Fibonacci number $F_n$ and write a new function, `fibonacci2(n)`.

```python
def fibonacci2(n):
    f1 = 1; f2 = 1; f = 1
    for i in range(0, n-2):
        f = f1 + f2
        f1 = f2; f2 = f
    return f
```

```python
[fibonacci(i) for i in range(1, 11)]
```

#### Q 1.1.4
What's the complexity of your implementation?

$T(n) = \mathcal{O}(n)$ in this case because the for loop over roughly $n$ times.

### Q 2.3: Is normality uniform?

![normal and uniform distributions](https://qph.ec.quoracdn.net/main-qimg-f6ed71ed1d0059760fb63db384dcbcca-c)

Given draws from a normal distribution with known parameters, how can you simulate draws from a uniform distribution?

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
%matplotlib nbagg
```

```python
seed = 42
n_samples = 500
```

- Let $X \sim \mathcal{N}(0,1)$ and $Y \sim \mathcal{U}(0,1)$ where $\mathcal{U}(0,1)$ is the uniform distribution over the inteval $[0,1]$.

```python
X = stats.norm
Y = stats.uniform
```

- draw IID samples from $\mathcal{N}(0,1)$ denoted by $X_i$

```python
xi = X.rvs(loc=0, scale=1, size=n_samples, random_state=seed)
xi[0:10]
```

```python
domain_normal = np.linspace(X.ppf(0.001), X.ppf(0.999)) 
```

- plot the PDF of $X$ along with the historgram of the sample $(x_1, x_2,\dots,x_n)$

```python
fig, ax = plt.subplots(1)
ax.plot(domain_normal, X.pdf(domain_normal), label='PDF')
ax.hist(xi, density=True, label='histogram of sample')
ax.legend()
```

- this verifies that the sample $(x_1, x_2,\dots,x_n)$ is drawn from a normal distribution

- Let $y_i = F_X(x_i)$, where $F_X$ is the CDF of $X$
- Create sample $(y_1, y_2,\dots, y_n)$

```python
yi = X.cdf(xi)
yi[0:10]
```

```python
domain_uniform = np.linspace(Y.ppf(0.001), Y.ppf(0.999))
```

- plot the historgram of the sample $(y_1, y_2,\dots, y_n)$ along with the uniform distribution for $Y$

```python
fig, ax = plt.subplots(1)
ax.plot(domain_uniform, Y.pdf(domain_uniform), label='PDF')
density, bins, _ = ax.hist(yi, density=True, label='historgram of sample')
ax.legend()
```

- we see that $(y_1, y_2,\dots, y_n)$ seem to be sampled from the uniform distribution.
- we have transformed a sample drawn from a normal distribution into a sample drawn from a uniform distribution.

Why does this work? Let us compute the CDF of $W=F_X(X)$, denoted by $F_W$. It is clear that the range of $W$ is between 0 and 1.

$F_W(w) = \mathbb{P}(W \leq w) = \mathbb{P}(F_X(X) \leq w) = \mathbb{P}(\{x, F_x(x) \leq w\})$ 
        = $\mathbb{P}(\{x, x \leq F^{-1}(w)\})$ 
        
$\mathbb{P}(\{x, x \leq F^{-1}(w)\}) = \int_{-\infty}^{F^{-1}_X(w)} dF_X(x) = F_X(F^{-1}_X(w)) = w$

Therefore $F_W(w) = w$ which is the CDF of a uniform distribution. So $W, Y$ are equal in the distribution sense.
